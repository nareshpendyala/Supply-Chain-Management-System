```python
import redis
import pandas as pd
from datetime import datetime
import time
r = redis.Redis(host='localhost', port=7000, db=0)
r.sadd('Partition-1', "Supplier-1", "Supplier-2", "Supplier-3" , "Supplier-4")
r.sadd('Partition-2', "Manufacturer-1", "Manufacturer-2", "Manufacturer-3" , "Manufacturer-4")
r.sadd('Partition-3', "Distributer-1", "Distributer-2", "Distributer-3" , "Distributer-4")
r.sadd('Partition-4', "Retailer-1", "Retailer-2", "Retailer-3" , "Retailer-4")
r1 = redis.Redis(host='localhost', port=7001, db=0)
r2 = redis.Redis(host='localhost', port=7002, db=0)
r3 = redis.Redis(host='localhost', port=7003, db=0)
r4 = redis.Redis(host='localhost', port=7004, db=0)
```

```python
df = pd.read_csv("Incoming Messages data.csv")
for row in df.iterrows():
    mess = (row[1]).to_dict()
    t = datetime.now().strftime('%H:%M:%S')
    if r.sismember("Partition-1",mess['entitiy']) : out_port = r1
    elif r.sismember("Partition-2",mess['entitiy']) : out_port = r2
    elif r.sismember("Partition-3",mess['entitiy']) : out_port = r3
    elif r.sismember("Partition-4",mess['entitiy']) : out_port = r4
    print("A message from", mess['entitiy'] ," received & distributed ...")
    out_port.rpush("Messages", str(t) + "--" + str( mess['value']))
    time.sleep(0.01) 
```
