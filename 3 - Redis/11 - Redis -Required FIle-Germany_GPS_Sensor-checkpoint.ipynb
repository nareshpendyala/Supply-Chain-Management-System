-: Redis User Story :- 1 

A Transportation customer service desk employee got a call from End user, he is asking about status of his delivery. The employee checked the information in his office. That perticular piece is out for delivery. So, it is in between parcel delivery center & End user home.

To give answer to enduser, employee need real time tracking infomation ?

```python
import pandas as pd
import time
df = pd.read_csv("Germany Location data.csv")
```

```python
import redis
r = redis.Redis(host='localhost', port=8000, db=0)
for row in df.iterrows():
    r.hmset("loc_ger",(row[1]).to_dict())
    print("Generating Germany Data.....")
    time.sleep(3)
```
